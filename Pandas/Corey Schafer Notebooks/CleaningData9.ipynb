```python
import pandas as pd
import numpy as np
```

```python
people={
    'first':['Corey','Jane','John','Chris', np.nan, None, 'NA'],
    'last':['Schafer','Doe','Doe','Schafer', np.nan, np.nan, 'Missing'],
    'email':['CoreyMSchafer@gmail.com','JaneDoe@email.com','JohnDoe@email.com',None, np.nan,'Anonymous@email.com','NA'],
    'age':['33','55','63','36',None,None,'Missing']    
}
```

```python
df=pd.DataFrame(people)
```

```python
df
```

```python
#To delete row with NA value
df.dropna()
```

```python
#In above command the default values are
df.dropna(axis='index', how='any')  #for any if row contains any one of the cell data as None or nan then the row get deleted


#axis value can be index or columns  if columns then column get drop
#how values can be any or all if all then row get deleted only if all the element in the row are None or nan
```

```python
df.dropna(axis='index', how='all')  #for all if all cell in a row contains cell data as None or nan then only the row get deleted


#axis value can be index or columns  if columns then column get drop
#how values can be any or all if all then row get deleted only if all the element in the row are None or nan
```

```python
df.dropna(axis='columns', how='all')
```

```python
df.dropna(axis='columns', how='any')  #clear all because in each of the column there is one nan or None element
```

```python
#To print data of perticular column of no data in that column then drop that row
df.dropna(axis='index', how='any', subset=['email'])   #here are applying how condition only on subset email column
```

```python
df.dropna(axis='index', how='all', subset=['email']) 
```

```python
#To pass multiple column    If data in both last and email is none or nan then only delete the rows
df.dropna(axis='index',how='all', subset=['last','email'])
```

```python
df.dropna(axis='index',how='any', subset=['last','email']) # delete row if any one of the column contain None or nan
```
