```python
from matplotlib import pyplot as plt
```

```python
#Pie Chart is better for small data
plt.style.use("fivethirtyeight")
slices=[30,60,100,70,50]
labels=['Thirty','Sixty','Hundred','Seventy','Fifty']
colors=['green','#2211FF','#112233','#444455','#881111']
plt.pie(slices, labels=labels,colors=colors,wedgeprops={'edgecolor':'black'})
plt.title("Pie Chart")
plt.tight_layout() 
plt.show()
```

```python
#Pie Chart is better for small data
plt.style.use("fivethirtyeight")
slices=[30,60,100,70,50]
explode=[0,.1,0,.1,0] #bigger the no. bigger the impact
labels=['Thirty','Sixty','Hundred','Seventy','Fifty']
colors=['green','#2211FF','#112233','#444455','#881111']
plt.pie(slices, labels=labels,colors=colors,explode=explode,shadow=True,wedgeprops={'edgecolor':'black'})
plt.title("Pie Chart")
plt.tight_layout() 
plt.show()

```

```python
#Pie Chart is better for small data
plt.style.use("fivethirtyeight")
slices=[30,60,100,70,50]
explode=[0,.1,0,.1,0] #bigger the no. bigger the impact
labels=['Thirty','Sixty','Hundred','Seventy','Fifty']
colors=['green','#2211FF','#112233','#444455','#881111']
plt.pie(slices, labels=labels,colors=colors,explode=explode,shadow=True,
        startangle=90,
        autopct="%1.1f%%",#To print percentage
        wedgeprops={'edgecolor':'black'})
plt.title("Pie Chart")
plt.tight_layout() 
plt.show()
```
