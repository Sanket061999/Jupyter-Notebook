```python
with open('data.csv') as csv_file:
    csv_reader=csv.DictReader(csv_file)
    language_counter=Counter()
    for row in csv_reader:
        language_counter.update(row['LanguagesWorkedWith'].split(';'))
#print(language_counter) will return dict
#print(language_counter.most_common(15)) will return list of tuple

languages=[]
popularity=[]
for item in language_counter.most_common(15):
    languages.append(item[0])
    popularity.append(item[1]) 
#print(languages)
#print(popularity)
languages.reverse()
popularity.reverse()

plt.barh(languages,popularity) #when we have large no. of bars then we can use horizontal bar
plt.ylabel('Languages')
plt.xlabel('No of User')
plt.title("User vs Languages")
plt.grid(False)
plt.show()
```

```python
#Using Pandas
data=pd.read_csv('data.csv')
lang_responses=data['LanguagesWorkedWith']
for response in lang_responses:
    language_counter.update(response.split(';'))


languages=[]
popularity=[]
for item in language_counter.most_common(15):
    languages.append(item[0])
    popularity.append(item[1]) 
#print(languages)
#print(popularity)
languages.reverse()
popularity.reverse()

plt.barh(languages,popularity,color='b') #when we have large no. of bars then we can use horizontal bar
plt.ylabel('Languages')
plt.xlabel('No of User')
plt.title("User vs Languages")
plt.grid(False)
plt.show()

    
```
